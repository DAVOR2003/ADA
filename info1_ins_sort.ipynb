# 1-PROBLEMA DE ORDENAMIENTO (INSERTION SORT)


**¿EN QUE CONSISTE EL INSERTION SORT?**\
Consiste que en una lista o arreglo se recibe de **ENTRADA**: $[a_1,a_2,..,a_n]$ numeros , y este va recibiendo número a número comparando el numero ingresado con los ya recibidos y cada vez que en numero sea menos al que este se va corriendo hacia atras y asi sucesivamente hasta lograr que **SALIDA**: $[a'_1<a'_2<....<a'_n]$

#2-INSERTION SORT


el siguente codigo muestra una implementacion den **insertion sort**:

#2.1 código

```python
import random
from termcolor import colored

def insertion_sort(a,verbose=False):
   n = len(a)
   t=0
   if verbose == True: print("input array:",a)
   for i in range(1,n):
     if verbose == True: print("\nPasada",i)
     already_sorted = True
     key = a[i]
     j = i-1
     while (j >=0) and (key < a[j]) :
              a[j+1] = a[j]
              j -= 1
              t=t+1
    
     a[j+1] = key
     if j>=0 : t = t+1
     if verbose == True: 
                   print(str(a[:i])[1:-1],",",colored(f"{str(a[i:i+2])[1:-1]}","red"),",",str(a[i+2:])[1:-1])
    
   return a,t



A = [5,8,7,1,3]
print("Entrada: ",A)
A, counter = insertion_sort(A)
print("Salida: ",A)
print("# comparaciones: ", counter)


```

#2.2 Descripcion insertion sort

el insertion sort consistira en resivir un arreglo A que contendran N cantidad de numeros y este los ordenara de menos a mayor:\
**paso a paso**:


1-se crea un ciclo iterativo con el largo del arreglo\
2-se copia el valor a comparar en la variable **key**\
3-se compara el valor de **key** con los valores antes de este para poder ver si es que es menor a uno de estos y poder saber en que lugar debe ubiucarse este numero\
4-cuando nse verifique el lugar en el que debe ubicarse se coloca en el lugar y se repite esto sucesivamente

#2.3 Ejemplo:

condireremos el arreglo:\
a = [3,2,1,4,5]

en la primera pasada cambiamos el 2 con el 3:\
a=[2,3,1,4,5]

en la segunda pasada se cambia el 1 cion el 2 y el 3 :\
a=[1,2,3,4,5]

y el codigo se iterara sin cambiar nada hasta que termine finalmente y queda el arreglo ordenado.

#2.4 Ejecución del algoritmo paso a paso (verbose=True)

```python
import random
a = random.sample(range(1, 100), 6)
a,c= insertion_sort(a,verbose=True)
```

#3-Tiempo de ejecucion

**Teorema (tiempo de ejecucion)**

El algoritmo insertion sort tiene un tiempo de ejecución de O(n^2)
 en el peor caso.

**Prueba del teorema**

Observando el código, podemos apreciar que el peor caso ocurre cuando el menor elemento se encuentra al final del arreglo. En este caso, en cada pasada el elemento se movería una casilla a la izquierda, por lo que se necesitaría realizar **n-1** pasadas para que los elementos quedaran ordenados.

**EJEMPLO**

```python
a =[2,3,4,1];
print("Entrada no ordenada:",a)
a,counter=insertion_sort(a,True)
print("Salida ordenada:",a)
print("Total de comparaciones realizadas:",counter)
```

en este caso en la primera pasada comparo 1 vez ,la segunda 1 vez nuevamente y a la tercera 3 veces.

En el caso general, se deberían realizar $\sum\limits_{i=0}^{n-1} i=\frac{n(n-1)}{2}$ comparaciones. Por lo que el **tiempo de ejecución del algoritmo en el peor caso** es $O(n^2)$. 

#4-CORRECTITUD


### **Teorema (Correctitud).**

*El algoritmo **InsertionSort** genera un arreglo: [a_1',a_2',...,a_n'] con los mismos elementos del arreglo de entrada ordenados de menor a mayor, es decir, * $a_1'\leq a_2' \leq... \leq a_n'$.

## Prueba del Teorema

Previamente observamos que en cada iteración de InsertionSort se van comparando elementos de izquierda a derecha así mismo acomodando los elementos.

Por lo que probaremos la siguiente **propiedad invariante de bucle** para el `InsertionSort`:

**Inicialización**

Al comienzo de la primera iteración (j=1), los primeros elementos cumplen con la propiedad.

**Mantención**

Aplicando inducción, asumimos que al comienzo de cada iteración $j$ se cumple la propiedad. Es decir, los primeros elementos de izquierda a derecha van quedando ordenados de menor a mayor.

image.png

Como se observa a medida que **$j$** va aumentando los elementos de izquierda a derecha quedan ordenados.

**Correctitud**

Finalmente, como la propiedad de bucle invariante es verdadera al incio del bucle y se mantiene en cada iteración, podemos decir que al **finalizar la n-ésima iteración del algoritmo**, se que genera un arreglo $[a_1',a_2',...,a_n']$, con los mismos elementos del arreglo de entrada ordenados de menor a mayor. $\Box$

5.1 Número de comparaciones

A continuación se muestra gráficamente una comparación entre:

* cantidad de comparaciones del peor caso calculadas matemáticamente,
* cantidad de comparaciones del mejor caso calculadas matemáticamente (n-1 comparaciones si el arreglo está ordenado)
* cantidad de comparaciones realizadas experimentalmente para entradas aleatorias

para tamaños de problemas con $n \in [5,19]$.

Los arreglos de entrada son generados aleatoriamente haciendo un sampling de un conjunto de números entre 1 y 100.

```python
import matplotlib.pyplot as plt

x=[n for n in range(5,20)] 
y1=[n*(n-1)/2 for n in range(5,20)] # worst case
y2=[n-1 for n in range(5,20)] # best case
y=[]; 

for n in range(5,20):
  a = random.sample(range(1, 100), n)
  a,counter = insertion_sort(a)
  y.append(counter)

plt.plot(x,y)
plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(["insertion sort", "theoretical worst case", "theoretical best case"])

plt.xlabel('n')
plt.ylabel('number of operations')
plt.show()
```

#Análisis resultados

como podemos observar el grafico creado los casos realizados aleatoriamente esta entremedio del mejor y del peor casi pero ligeramente inclinado hacia el peor caso.
