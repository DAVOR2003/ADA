#1-Problema de ordenamento (Merge Sort)

El algoritmo **MergeSort** es de tipo divide y vencerás ya que aplica los 3 pasos del paradigma. Al igual que InsertionSort, el algoritmo MergeSort recibe una secuencia de números como entrada y retorna la secuencia ordenada de menor a mayor.

**Entrada**: Secuencia de n números $[a_1,a_2,...,a_n]$

**Salida**: Permutación ordenada de la secuencia de entrada: $[a_1',a_2',...,a_n']$, de tal manera que $a_1'\leq a_2' \leq... \leq a_n'$.

 # 2.1 Codigo:

```python
import random
from termcolor import colored
import copy 
import termcolor 

def mergeSort(arr,verbose=False):
    if len(arr) > 1:
        mid = len(arr)//2
        L = arr[:mid]
        R = arr[mid:]
        global cont
        if verbose == True :
          termcolor.cprint(f"parte izquierda: {L} ",'red')
          termcolor.cprint(f"parte derecha: {R} ",'blue')
          print("\n")
        cont +=1
        mergeSort(L,verbose)
        cont+=1
        mergeSort(R,verbose)
        i = j = k = 0

        while i < len(L) and j < len(R):
            global comp
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1
            comp+=1
        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1
        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1      
        if verbose == True :
           termcolor.cprint(f"parte izquierda: {L} ",'red')
           termcolor.cprint(f"parte derecha: {R} ",'blue')
           termcolor.cprint(f"arreglo ordenado: {arr} ",'magenta')
           print("comparaciones: ",comp)
           print("\n")

def printList(arr):
    for i in range(len(arr)):
        print(arr[i], end=" ")
    print()
 

arr = [3,7,8,9,2,4,6]
cont = comp = 0
print("arreglo recibido", end="\n")
printList(arr)
verbose = False
mergeSort(arr,verbose)
print("arreglo de salida", end="\n")
printList(arr)
print("llamadas recursivas:",cont)
print("cantidad comparaciones:",comp)
  

 
```

#2.2 Descripcion algoritmo(Merge Sort):

paso a paso como funciona el algoritmo:

1. Divide la secuencia en dos subsecuencias de n/2 elementos cada una.
2. Ordena cada subsecuencia de manera recursiva usando MergeSort.
3. Combina (merge) las dos subsecuencias ordenadas para obtener una secuencia ordenada de n números.\
como funciona el merge:\
4. Creamos un nuevo arreglo B para colocar los elementos ordenados.
5. Comparamos el primer elemento de cada sub-arreglo A y colocamos el menor de ellos en B. Eliminamos el elemento seleccionado del sub-arreglo A correspondiente.
6. Repetimos el paso 1 hasta que uno de los sub-arreglos A quede vacío.
7. Agregamos los elementos del otro arreglo directamente en B
8. Copiamos el contenido de B en A[p..r].

#4.1-Teorema de correctitud:


*El algoritmo **MergeSort** genera un arreglo: $[a_1',a_2',...,a_n']$, con los mismos elementos del arreglo de entrada ordenados de menor a mayor, es decir,* $a_1'\leq a_2' \leq... \leq a_n'$.

Para probar la correctitud de **algoritmos recursivos** usaremos **inducción matemática**.

En **inducción matemática,** para probar una declaración $P(n)$, para todos los números naturales $n$, típicamente debemos:

1. Probar $P(n)$ para un **caso base**, por ejemplo $P(1)$
2. Probar que si $P(m)$ es cierto para $m<n$, entonces $P(n)$ también lo es.

Nuestra **hipótesis $P(n)$** es que MergeSort, retorna un **arreglo** con la secuenciua de $n$ elementos de entrada ordenados de menor a mayor. 

En el **caso base**, es decir, cuando $n=1$, MergeSort retorna el único elemento, por lo que se cumple $P(1)$.

Para el **paso inductivo**, debemos probar que si $P(m)$ es cierta para $m<n$, entonces también lo es $P(n)$. 

Para $n$ elementos, el algoritmo divide el arreglo en dos subarreglos de tamaños: $m=\lfloor \frac{n}{2}\rfloor$ y $m=\lceil \frac{n}{2}\rceil$ y las ordena usando **Mergesort**. Como el tamaño de los subarreglos es menor a $n$, con el paso inductivo asumimos que son ordenadas correctamente. 

# Luego, los dos subarreglos **son combinados con la función Merge**. Según el Teorema 1, Merge retorna un arreglo ordenado con los mismos elementos de los subarreglos, por lo tanto **MergeSort es correcto**. 

#4.2 prueba para tiempo de ejecucion(arbol recursivo)

Sin usar ningun tipo de teorema, podemos intuir por qué la complejidad de MergeSort es $O(n \log n)$. Primero, reemplazemos $O(1)$ por una constante $c$ y las complejidades $O(n)$ por el término $c \cdot n$.

$T(n) = \begin{cases}
  c  & n \text{ is small} \\
  2 T(n/2) + c \cdot n & \text{ otherwise}
\end{cases}$

Luego, vamos subdividiendo con uno a uno , es decir: $T(n)=2T(n/2)+cn$.


Hasta llegar al siguiente árbol expandido completamente:

Analizando el árbol nos damos cuenta que la suma de tiempos de cada nivel es $cn$. 

Por otro lado, la altura del árbol es $\log_2 n$.

Finalmente, la suma total de tiempos de la recursión es: $cn \log_2 n \rightarrow O(n \log n)$
