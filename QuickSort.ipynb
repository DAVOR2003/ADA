#1-Descripci√≥n del problema,algoritmo(QuickSort):

Este algoritmo utiliza el metodo conocido como divide y venceras, lo que sucede es que recibe un arreglo: \
Largo N: \
Entrada:[a1,a2,...,an] \
y este lo que hace es seleccionar un indice es decir un elemento del algoritmo y lo que ocurre en el algoritmo es el arreglo dividirlo en base al elemento elegido, se mueven los elementos menores al elegido hacia la izquierda y los mayores a este a la derecha, y esto va sucediendo por cada parte del algoritmo hasta mostrar una salida con los elementos ordenados \
Salida:[a'1,a'2,....,a'n] \

```python
#pivote como primer elemento
import random
from termcolor import colored
import termcolor
global comp
global verbose

def partitionA(arr, l, h):
  global comp
  global verbose
  low, high = l, h
  if l != h and l < h:
    pivot = arr[l]
    low = low+1
  
    while low <= high:
      comp=comp+2
      if arr[high] < pivot and arr[low] > pivot:
        arr[high], arr[low] = arr[low], arr[high]
        if verbose == True :
          print(colored("numeros a cambiar:","green"),arr[high], arr[low])

      if not arr[low] > pivot:
        low += 1
      if not arr[high] < pivot:
        high -= 1
  if verbose == True :
        print(arr)   
  if verbose == True :
     print(colored("numeros a cambiar:","green"),arr[high], arr[l])   
  arr[l], arr[high] = arr[high], arr[l]
  if verbose == True :
        print(arr) 
  return high
  
def quick_sortA(array, low, high):
  if low < high:
      if verbose==True :
        print(colored("numero de particion:","yellow"),low)
      pi = partitionA(array, low, high)
      quick_sortA(array, low, pi - 1)
      quick_sortA(array, pi + 1, high)
  
  

comp=0
verbose=False
n = random.randint(4,10)
array = random.sample(range(-100,100),n)
print(colored("arreglo desordenado:","red"))
termcolor.cprint(f"{array}",'red')
print("")
quick_sortA(array, 0, len(array) - 1)
print(colored("arreglo ordenado:","red"))
termcolor.cprint(f"{array}",'red')
print(colored("comparaciones:","blue"),comp)
```

#2.2-Descripci√≥n del algoritmo(QuickSort):

paso a paso como funciona el QuickSort: \


1-Se elge el elemento del arreglo por el cual se quiere trabajar. \
2-el elemento se meve al final del arreglo y el arreglo se va ordenando de forma que a la parte izquierda se pondran los elementos menores y a la derecha lso mayores. \
3- esto se va haciendo recursivamente con cada arreglo izquierda y derecha hasta poder lograr que al junbtar ambos arreglos esten de forma ordenada. 


#2.3-Ejemplo:

consideremos la lista: \
A = [ 2 , -10 , 6 , 10 , 4 ] \
tomaremos los ultimos elementos como indices: \
L = [ 2 , -10 ] M = [ 4 ] R =[ 6 , 10 ] \
el lado R ya esta ordenado asi que ordenaremos el izquierdo: \
L' = [ -10 ] L'' = [ 2 ] M = [ 4 ] R = [ 6 , 10 ]  \
asi que finalizando juntamos todo y se obtendra el arreglo ordenado: \
A' = [ -10 , 2 , 4 , 6 , 10 ]

```python
import random
from termcolor import colored
import copy 
import termcolor 
global comparaciones

def mergeSort(arra,verboses=False):
    if len(arra) > 1:
        mid = len(arra)//2
        L = arra[:mid]
        R = arra[mid:]
        if verbose == True :
          termcolor.cprint(f"parte izquierda: {L} ",'red')
          termcolor.cprint(f"parte derecha: {R} ",'blue')
          print("\n")
        mergeSort(L,verboses)
        mergeSort(R,verboses)
        i = j = k = 0

        while i < len(L) and j < len(R):
            global comparaciones
            if L[i] < R[j]:
                arra[k] = L[i]
                i += 1
            else:
                arra[k] = R[j]
                j += 1
            k += 1
            comparaciones=comparaciones+1
        while i < len(L):
            arra[k] = L[i]
            i += 1
            k += 1
        while j < len(R):
            arra[k] = R[j]
            j += 1
            k += 1      
        if verboses == True :
           termcolor.cprint(f"parte izquierda: {L} ",'red')
           termcolor.cprint(f"parte derecha: {R} ",'blue')
           termcolor.cprint(f"arreglo ordenado: {arra} ",'magenta')
           print("comparaciones: ",comparaciones)
           print("\n")

def printList(arr):
    for i in range(len(arr)):
        print(arr[i], end=" ")
    print()
 

arra = [3,7,8,9,2,4,6]
comparaciones = 0
print("arreglo recibido", end="\n")
printList(arra)
verboses = False
mergeSort(arra,verboses)
print("arreglo de salida", end="\n")
printList(arra)
print("cantidad comparaciones:",comparaciones)
  
```
